```python
import matplotlib.pyplot as plt
import pandas as pd

# Cargar los datos (si no los tienes ya)
df = pd.read_csv('/content/simplificad.csv.txt')

# Configuración básica
plt.figure(figsize=(15, 10))

# 1. Gráfico de PIB
plt.subplot(2, 2, 1)
for pais in ['el salvador', 'mexico']:
    datos = df[df['pais'] == pais]
    plt.plot(datos['anio'], datos['pib_per_capita_usd'], label=pais, marker='o')
plt.title('Evolución del PIB per cápita')
plt.xlabel('Año')
plt.ylabel('PIB (USD)')
plt.legend()

# 2. Gráfico de Internet y Smartphones
plt.subplot(2, 2, 2)
for pais in ['el salvador', 'mexico']:
    datos = df[df['pais'] == pais]
    plt.plot(datos['anio'], datos['penetracion_internet'], label=f'{pais} - Internet', linestyle='-', marker='o')
    plt.plot(datos['anio'], datos['uso_smartphones'], label=f'{pais} - Smartphones', linestyle='--', marker='s')
plt.title('Penetración de Internet vs Smartphones')
plt.xlabel('Año')
plt.ylabel('Porcentaje (%)')
plt.legend()

# 3. Gráfico de Bancarización
plt.subplot(2, 2, 3)
for pais in ['el salvador', 'mexico']:
    datos = df[df['pais'] == pais]
    plt.plot(datos['anio'], datos['tasa_bancarizacion'], label=pais, marker='o')
plt.title('Tasa de Bancarización')
plt.xlabel('Año')
plt.ylabel('Porcentaje (%)')
plt.legend()

# 4. Gráfico de Criptomonedas
plt.subplot(2, 2, 4)
for pais in ['el salvador', 'mexico']:
    datos = df[df['pais'] == pais]
    plt.plot(datos['anio'], datos['uso_criptomonedas'], label=pais, marker='o')
plt.title('Uso de Criptomonedas')
plt.xlabel('Año')
plt.ylabel('Porcentaje (%)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Datos hipotéticos basados en información real
datos_ampliados = [
    {
        'pais': 'México',
        'anio': 2023,
        'volumen_transacciones_cripto': 2500,  # en millones USD
        'negocios_aceptan_cripto': 1250,
        'adopcion_ciudadana': 12.5,  # % de población
        'crecimiento_anual': 8.2  # %
    },
    {
        'pais': 'El Salvador',
        'anio': 2023,
        'volumen_transacciones_cripto': 450,  # en millones USD
        'negocios_aceptan_cripto': 2300,
        'adopcion_ciudadana': 24.7,  # % de población
        'crecimiento_anual': 15.3  # %
    }
]

df_ampliado = pd.DataFrame(datos_ampliados)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
sns.set_style("whitegrid")
plt.figure(figsize=(14, 8))

# Gráfico 1: Adopción y crecimiento
plt.subplot(2, 2, 1)
sns.barplot(x='pais', y='adopcion_ciudadana', data=df_ampliado, palette='viridis')
plt.title('Adopción Ciudadana de Criptomonedas (%)')
plt.ylabel('Porcentaje de población')

# Gráfico 2: Volumen de transacciones
plt.subplot(2, 2, 2)
sns.barplot(x='pais', y='volumen_transacciones_cripto', data=df_ampliado, palette='magma')
plt.title('Volumen de Transacciones (Millones USD)')
plt.ylabel('Millones de dólares')

# Gráfico 3: Negocios que aceptan cripto
plt.subplot(2, 2, 3)
sns.barplot(x='pais', y='negocios_aceptan_cripto', data=df_ampliado, palette='rocket')
plt.title('Negocios que Aceptan Criptomonedas')
plt.ylabel('Cantidad de negocios')

# Gráfico 4: Crecimiento anual
plt.subplot(2, 2, 4)
sns.barplot(x='pais', y='crecimiento_anual', data=df_ampliado, palette='mako')
plt.title('Crecimiento Anual de Adopción (%)')
plt.ylabel('Porcentaje de crecimiento')

plt.tight_layout()
plt.show()
```
